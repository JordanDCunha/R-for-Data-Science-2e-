## â–¶ï¸ **6.1.1 Running Code**

The most important shortcut in RStudio:

> **Cmd/Ctrl + Enter**

This runs the **current R expression** in the console.

When your cursor is inside a multi-line expression, pressing Cmd/Ctrl + Enter:
- Executes the full expression
- Moves the cursor to the next line
- Makes it easy to step through a script line-by-line

You can also run the **entire script at once** using:

> **Cmd/Ctrl + Shift + S**

Running your whole script regularly helps ensure everything needed for your analysis is captured.

## ðŸ’¾ **6.1.3 Saving and Naming Scripts**

RStudio auto-saves script contents, but you should still:
- **Save scripts explicitly**
- **Avoid names like** `Untitled1.R`, `myscript.R`, or `code.R`

### ðŸ·ï¸ File Naming Principles

Good file names should be:

1. **Machine-readable**
   - No spaces or special characters
   - Avoid relying on capitalization

2. **Human-readable**
   - Describe what the file contains

3. **Well-ordered**
   - Start with numbers so files sort logically

---

### âŒ Poor File Names

- Ambiguous purpose
- Confusing capitalization
- No clear execution order

### âœ… Better Organization

- Descriptive names clarify purpose

If your project grows, go one step further:

Sooner or later, real life will interrupt your R session. Youâ€™ll quit R, work on something else, juggle multiple analyses, import data from outside R, and export results back out. To handle all of this smoothly, you need to answer two key questions:

## ðŸ§  **6.2.1 What Is the Source of Truth?**

As a beginner, itâ€™s fine to rely on the **Environment** to hold objects youâ€™ve created. But for larger projects or collaboration, your **R scripts** should be the source of truth.

Why?
- Scripts + data files can **recreate your environment**
- An environment alone cannot reliably recreate your scripts
- Relying on memory or R history leads to mistakes

To enforce this discipline, itâ€™s strongly recommended that RStudio **does not restore your workspace between sessions**. You can do this by:
- Running `usethis::use_blank_slate()`
- Or manually setting RStudio options to always start clean

This causes short-term inconvenience (nothing is remembered on restart), but it prevents long-term disastersâ€”like realizing months later that a critical calculation exists only as an object, not as code.

### ðŸ” A Powerful Workflow Pattern
Use these two shortcuts together:
1. **Cmd/Ctrl + Shift + 0 (or F10)** â†’ Restart R
2. **Cmd/Ctrl + Shift + S** â†’ Re-run the entire script

This ensures your analysis is fully reproducible from scratch.

### âš ï¸ Note on RStudio Server
RStudio Server **does not restart R by default** when you close the browser tab. This makes it even more important to manually restart R regularly so youâ€™re not unknowingly relying on leftover objects.

## ðŸ“ **6.2.2 Where Does Your Analysis Live?**

R uses a **working directory**, which is:
- Where files are loaded from
- Where outputs are saved

RStudio shows this at the top of the console, and you can print it with `getwd()`.

As a beginner, itâ€™s okay if your working directory is messy. But at this stage, you should:
- Organize work into **project-specific directories**
- Set the working directory to the project folder

You *can* change the working directory manually using `setwd()`, but this is **not recommended**. Thereâ€™s a better solution.

---

## ðŸ“¦ **6.2.3 RStudio Projects**

An **RStudio project** keeps everything related to an analysis together:
- R scripts
- Data
- Figures
- Outputs

Create one via:
**File â†’ New Project â†’ New Directory â†’ New Project**

Choose a sensible location and name the project carefullyâ€”it should be easy to find later.

Once created:
- The project directory becomes the working directory
- Opening the `.Rproj` file restores your workspace layout
- Your environment starts fresh (if configured correctly)

This guarantees reproducibility and organization.

A major benefit:
If you save figures and data **only via code**, you can always trace outputs (like PNGs) back to the scripts that created them.

---

## ðŸ§­ **6.2.4 Relative vs Absolute Paths**

Inside a project, **always use relative paths**.

- **Relative paths** are based on the projectâ€™s root  
  Example: `data/diamonds.csv`
- **Absolute paths** depend on your computerâ€™s exact folder structure  
  Example: `/Users/name/Documents/project/data/diamonds.csv`

Relative paths:
- Work on any computer
- Make collaboration possible
- Keep projects portable

Absolute paths:
- Break when shared
- Should never appear in scripts

### ðŸ–¥ï¸ Path Separators
- Mac/Linux: `/`
- Windows: `\`

R supports both, but backslashes must be escaped (`\\`), which is annoying. Best practice: **always use forward slashes**.
