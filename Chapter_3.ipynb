# âœ¨ **3.2 Rows** âœ¨

The most important **dplyr verbs** that operate on **rows** of a dataset are:

- **`filter()`** â†’ changes **which rows** are present without changing their order  
- **`arrange()`** â†’ changes the **order of rows** without changing which are present  
- **`distinct()`** â†’ keeps rows with **unique values**  

These functions affect **rows only**; columns are left unchanged unless explicitly requested.

---

## ðŸ”¹ **3.2.1 `filter()`**

**`filter()`** keeps rows based on logical conditions applied to columns.

- First argument: a **data frame**
- Remaining arguments: **conditions** that must be `TRUE`
- Output: a **new data frame**

Example: flights delayed more than two hours.

You can use comparison operators such as **`>`**, **`>=`**, **`<`**, **`<=`**, **`==`**, and **`!=`**.  
Combine conditions with **`&`** (and) or **`|`** (or).

A useful shortcut when combining `==` and `|` is **`%in%`**, which keeps rows where a variable equals any value in a vector.

To save filtered results, use the assignment operator **`<-`**, since dplyr never modifies inputs.

---

## âš ï¸ **3.2.2 Common Mistakes**

- Using **`=`** instead of **`==`** when testing equality  
- Writing â€œorâ€ statements like English (`month == 1 | 2`), which does not behave as intended  

---

## ðŸ”¹ **3.2.3 `arrange()`**

**`arrange()`** changes the order of rows based on column values.

- Multiple columns break ties in order
- The number of rows does **not** change

Use **`desc()`** to sort in descending order.

---

## ðŸ”¹ **3.2.4 `distinct()`**

**`distinct()`** finds unique rows in a dataset.

- With no arguments, it removes duplicate rows
- With selected columns, it finds unique combinations
- Use **`.keep_all = TRUE`** to retain all columns

`distinct()` keeps the **first occurrence** of each unique row.

To count occurrences instead, use **`count()`**, optionally sorting by frequency.

# âœ¨ **3.3 Columns** âœ¨

There are **four key dplyr verbs** that affect **columns** without changing the rows:

- **`mutate()`** â†’ creates **new columns** from existing ones  
- **`select()`** â†’ chooses **which columns** are present  
- **`rename()`** â†’ changes **column names**  
- **`relocate()`** â†’ changes the **order of columns**

---

## ðŸ”¹ **3.3.1 `mutate()`**

The purpose of **`mutate()`** is to **add new variables** that are calculated from existing variables.  
For now, we focus on **basic algebra**, such as computing:

- **gain** â†’ how much delay was made up in the air  
- **speed** â†’ miles per hour  

By default, new columns are added to the **right-hand side** of the data frame.

You can control placement using:
- **`.before`** â†’ place new variables before a column (or position)
- **`.after`** â†’ place new variables after a column (or position)

You can also control which variables are kept using **`.keep`**:
- `"used"` keeps only variables involved in the calculation and the newly created ones

Remember: unless assigned with **`<-`**, results are **not stored**.

---

## ðŸ”¹ **3.3.2 `select()`**

**`select()`** lets you focus on a subset of variables, which is especially helpful in large datasets.

You can:
- Select columns by **name**
- Select **ranges** of columns
- **Exclude** columns using `!`
- Select columns by **type**
- Use **helper functions** to match column names

Common helpers include:
- **`starts_with()`**
- **`ends_with()`**
- **`contains()`**
- **`num_range()`**

You can also **rename variables while selecting** using `new_name = old_name`.

---

## ðŸ”¹ **3.3.3 `rename()`**

Use **`rename()`** when you want to **keep all columns** but change the names of a few.

The syntax is:
- **new_name = old_name**

For large-scale cleaning, tools like **`janitor::clean_names()`** can help automate renaming.

---

## ðŸ”¹ **3.3.4 `relocate()`**

**`relocate()`** changes the **position of columns**.

- By default, selected columns move to the **front**
- Use **`.before`** or **`.after`** for precise placement
- Useful for grouping related variables or highlighting important ones

---

```python
# Create new columns with mutate()
flights |>
  mutate(
    gain = dep_delay - arr_delay,
    speed = distance / air_time * 60
  )

# Place new columns before all others
flights |>
  mutate(
    gain = dep_delay - arr_delay,
    speed = distance / air_time * 60,
    .before = 1
  )

# Place new columns after a specific column
flights |>
  mutate(
    gain = dep_delay - arr_delay,
    speed = distance / air_time * 60,
    .after = day
  )

# Keep only variables used in mutate
flights |>
  mutate(
    gain = dep_delay - arr_delay,
    hours = air_time / 60,
    gain_per_hour = gain / hours,
    .keep = "used"
  )

# Select specific columns
flights |>
  select(year, month, day)

# Select a range of columns
flights |>
  select(year:day)

# Exclude a range of columns
flights |>
  select(!year:day)

# Select character columns
flights |>
  select(where(is.character))

# Rename while selecting
flights |>
  select(tail_num = tailnum)

# Rename while keeping all columns
flights |>
  rename(tail_num = tailnum)

# Move columns to the front
flights |>
  relocate(time_hour, air_time)

# Relocate using helpers and position control
flights |>
  relocate(year:dep_time, .after = time_hour)

flights |>
  relocate(starts_with("arr"), .before = dep_time)
```

# âœ¨ **3.4 The Pipe (`|>`)** âœ¨

The **pipe operator** lets you **chain multiple data manipulation steps** together in a clear, readable sequence.  
Instead of nesting functions or creating many intermediate objects, the pipe passes the result of one step **directly** into the next.

---

## ðŸš€ **Why the Pipe Is Powerful**

With the pipe, you can **read code top to bottom** like a story:

1. Start with a dataset  
2. **Filter** the rows  
3. **Mutate** to create new variables  
4. **Select** the columns you care about  
5. **Arrange** the results  

Each verb appears at the **start of a new line**, making pipelines easy to skim and debug.

---

## ðŸ§© **Example: Fastest Flights to IAH**

Goal: Find the **fastest flights** going to **Houston (IAH)**.

Pipeline logic:
- Start with **`flights`**
- Keep only flights with destination `"IAH"`
- Compute **speed (mph)**
- Select relevant columns
- Sort by **descending speed**

---

## ðŸ”„ **Without the Pipe (Nested Calls)**

Without `|>`, code becomes **harder to read**, with functions buried inside other functions.

---

## ðŸ—‚ **Without the Pipe (Intermediate Objects)**

Another alternative is creating **multiple temporary objects**, which:
- Clutters the environment
- Makes workflows harder to follow

---

## âŒ¨ï¸ **Using the Pipe in RStudio**

- Insert the pipe with **Ctrl / Cmd + Shift + M**
- Make sure **â€œUse native pipe operator (`|>`)â€** is enabled in RStudio options

---

## ðŸ§  **Base Pipe vs magrittr Pipe**

Two pipe operators exist:

- **`|>`** â†’ Base R pipe (recommended)
- **`%>%`** â†’ magrittr pipe (tidyverse)

### Why prefer **`|>`**?
- Part of **base R** (always available)
- **Simpler** and more consistent
- Designed using lessons learned from `%>%`

For most everyday use, **both behave the same**, but `|>` is the modern standard.

---


```python
# Using the pipe to find the fastest flights to IAH
flights |>
  filter(dest == "IAH") |>
  mutate(speed = distance / air_time * 60) |>
  select(year:day, dep_time, carrier, flight, speed) |>
  arrange(desc(speed))

# The same logic without the pipe (nested functions)
arrange(
  select(
    mutate(
      filter(
        flights,
        dest == "IAH"
      ),
      speed = distance / air_time * 60
    ),
    year:day, dep_time, carrier, flight, speed
  ),
  desc(speed)
)

# The same logic using intermediate objects
flights1 <- filter(flights, dest == "IAH")
flights2 <- mutate(flights1, speed = distance / air_time * 60)
flights3 <- select(flights2, year:day, dep_time, carrier, flight, speed)
arrange(flights3, desc(speed))

# magrittr pipe example
library(tidyverse)

mtcars %>%
  group_by(cyl) %>%
  summarize(n = n())
```

# âœ¨ **3.5 Groups** âœ¨

So far, youâ€™ve worked with functions that operate on **rows** and **columns**.  
With **grouping**, **dplyr** levels up: you can now perform operations **within subsets of the data**.

The three most important tools here are:

- **`group_by()`** â†’ defines groups  
- **`summarize()`** â†’ reduces each group to summary statistics  
- **`slice_*()` family** â†’ extracts specific rows *within* each group  

---

## ðŸ§© **3.5.1 `group_by()`**

**`group_by()`** doesnâ€™t change the data itself â€” it changes **how future verbs behave**.

- It adds **grouping metadata**
- Subsequent operations happen **within each group**
- Youâ€™ll see grouping info printed at the top of the tibble output

Think of it as telling R:  
> â€œDo the next steps **separately for each group**.â€

---

## ðŸ“Š **3.5.2 `summarize()`**

**`summarize()`** collapses each group down to **one row** per group.

Key ideas:
- Missing values (`NA`) can break summaries  
- Use **`na.rm = TRUE`** to ignore missing data  
- You can compute **multiple summaries at once**
- **`n()`** counts rows per group (very useful!)

âš ï¸ Without `na.rm = TRUE`, many summaries will return **NA**.

---

## âœ‚ï¸ **3.5.3 The `slice_*()` Functions**

These functions let you **pick rows inside each group**:

- **`slice_head()`** â†’ first rows  
- **`slice_tail()`** â†’ last rows  
- **`slice_min()`** â†’ smallest values  
- **`slice_max()`** â†’ largest values  
- **`slice_sample()`** â†’ random rows  

Important detail:
- **Ties are kept by default**
- Use **`with_ties = FALSE`** if you want exactly one row per group

---

## ðŸ§® **3.5.4 Grouping by Multiple Variables**

You can group by **more than one column** (e.g., year + month + day).

When you summarize:
- dplyr **drops the last grouping level by default**
- A message explains what happened
- You can control this with **`.groups`**

Common options:
- **`"drop_last"`** â†’ default
- **`"drop"`** â†’ remove all grouping
- **`"keep"`** â†’ preserve all groups

---

## ðŸ”“ **3.5.5 `ungroup()`**

Use **`ungroup()`** when you want to:
- Remove grouping **without summarizing**
- Return to treating the data as **one big group**

After ungrouping:
- `summarize()` returns **a single row**

---

## ðŸ†• **3.5.6 The `.by` Argument**

Introduced in **dplyr 1.1.0**, **`.by`** allows **per-operation grouping**:

- No need for `group_by()` or `ungroup()`
- No grouping messages
- Cleaner for one-off summaries

This syntax is **experimental**, but very promising.

---

## ðŸ§  **Big Picture Takeaways**

- **Grouping changes behavior, not data**
- **Summaries answer â€œper-groupâ€ questions**
- **Slice functions return real rows, not statistics**
- **`.by` = modern, concise grouping**

Grouping + summaries = **most of real-world data analysis** ðŸš€


```python
# Grouping by month
flights |>
  group_by(month)

# Average departure delay by month (handling missing values)
flights |>
  group_by(month) |>
  summarize(
    avg_delay = mean(dep_delay, na.rm = TRUE)
  )

# Multiple summaries: mean delay + count
flights |>
  group_by(month) |>
  summarize(
    avg_delay = mean(dep_delay, na.rm = TRUE),
    n = n()
  )

# Using slice_max() to find the most delayed arrival per destination
flights |>
  group_by(dest) |>
  slice_max(arr_delay, n = 1) |>
  relocate(dest)

# Grouping by multiple variables
daily <- flights |>
  group_by(year, month, day)

# Summarizing grouped data (default: drop last group)
daily_flights <- daily |>
  summarize(n = n())

# Explicitly controlling grouping behavior
daily_flights <- daily |>
  summarize(
    n = n(),
    .groups = "drop_last"
  )

# Removing grouping
daily |>
  ungroup()

# Summarizing ungrouped data
daily |>
  ungroup() |>
  summarize(
    avg_delay = mean(dep_delay, na.rm = TRUE),
    flights = n()
  )

# Using the .by argument (no group_by needed)
flights |>
  summarize(
    delay = mean(dep_delay, na.rm = TRUE),
    n = n(),
    .by = month
  )

# .by with multiple grouping variables
flights |>
  summarize(
    delay = mean(dep_delay, na.rm = TRUE),
    n = n(),
    .by = c(origin, dest)
  )

# Tiny example data frame
df <- tibble(
  x = 1:5,
  y = c("a", "b", "a", "a", "b"),
  z = c("K", "K", "L", "L", "K")
)

# Inspect grouping
df |>
  group_by(y)

# Arrange rows
df |>
  arrange(y)

# Group + summarize
df |>
  group_by(y) |>
  summarize(mean_x = mean(x))

# Group by two variables
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x))

# Drop all groups explicitly
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x), .groups = "drop")

# Compare summarize vs mutate
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x))

df |>
  group_by(y, z) |>
  mutate(mean_x = mean(x))
```
